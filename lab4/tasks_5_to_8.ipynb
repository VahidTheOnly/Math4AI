```python
A = np.array([[1, 1], [2, 2]])
b = np.array([3, 6])
```

```python
U, S, VT = np.linalg.svd(A)
S_ = [[1/S[S>1e-5][0], 0], [0, 0]]
A_inv = VT.T @ S_ @ U.T
```

```python
A_inv @ b
```

# task 6

```python
import matplotlib.pyplot as plt
from face_data import Face1, Face2, Face3, TargetFace1, edges
```

```python
faces = np.vstack([Face1.flatten(), Face2.flatten(), Face3.flatten()]).T
Target_1 = TargetFace1.flatten()
```

```python
faces_first = faces[:3]
Target_1_first = Target_1[:3]
a, b, c = np.linalg.solve(faces_first.T @ faces_first, faces_first.T @ Target_1_first)

print(f'a = {a:0.1f}\nb = {b:0.1f}\nc = {c:0.1f}')

F = a * Face1 + b * Face2 + c * Face3

plt.scatter(TargetFace1[:, 0], TargetFace1[:, 1], color='r', s=50)
plt.scatter(F[:, 0], F[:, 1], color='g', s=40)
plt.xlim(-100, 100)
plt.ylim(-100, 100)
plt.show()
```
