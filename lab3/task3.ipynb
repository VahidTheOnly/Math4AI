```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib ipympl
```

```python
A = np.matrix([
    [1, 2, 1],
    [2, -1, -1],
    [-1, 1, -2]])

B = np.matrix([
    [1, 2, -3],
    [3, 1, 1],
    [2, 1, 0]])

C = np.matrix([
    [1, 2, -3],
    [3, 6, -9],
    [-2, -4, 6]])

u = np.random.rand(200, 3)

row_space_A = u @ A
col_space_A = A @ u.T

row_space_B = u @ B
col_space_B = B @ u.T


row_space_C = u @ C
col_space_C = C @ u.T
```

```python
fig1, ax1 = plt.subplots(2, 3, subplot_kw={'projection': '3d'}, figsize=(12, 8))
ax1[0, 0].scatter(col_space_A[0, :], col_space_A[1, :], col_space_A[2, :], color='r', edgecolors='black')
ax1[0, 0].set_title('A col space')
ax1[1, 0].scatter(row_space_A[:, 0], row_space_A[:, 1], row_space_A[:, 1], color='r', edgecolors='black')
ax1[1, 0].set_title('A row space')

ax1[0, 1].scatter(col_space_B[0, :], col_space_B[1, :], col_space_B[2, :], color='b', edgecolors='black')
ax1[0, 1].set_title('B col space')
ax1[1, 1].scatter(row_space_B[:, 0], row_space_B[:, 1], row_space_B[:, 2], color='b', edgecolors='black')
ax1[1, 1].set_title('B row space')

ax1[0, 2].scatter(col_space_C[0, :], col_space_C[1, :], col_space_C[2, :], color='g', edgecolors='black')
ax1[0, 2].set_title('C col space')
ax1[1, 2].scatter(row_space_C[:, 0], row_space_C[:, 1], row_space_C[:, 2], color='g', edgecolors='black')
ax1[1, 2].set_title('C row space')

plt.show()
```

```python
fig2, ax2 = plt.subplots(1, 3, subplot_kw={'projection': '3d'}, figsize=(12, 5))

ax2[0].scatter(col_space_A[0, :], col_space_A[1, :], col_space_A[2, :], color='r', edgecolors='black')
ax2[0].scatter(row_space_A[:, 0], row_space_A[:, 1], row_space_A[:, 2], color='b', edgecolors='black')
ax2[0].set_title('A row-col space')

ax2[1].scatter(col_space_B[0, :], col_space_B[1, :], col_space_B[2, :], color='r', edgecolors='black')
ax2[1].scatter(row_space_B[:, 0], row_space_B[:, 1], row_space_B[:, 2], color='b', edgecolors='black')
ax2[1].set_title('B row-col space')

ax2[2].scatter(col_space_C[0, :], col_space_C[1, :], col_space_C[2, :], color='r', edgecolors='black')
ax2[2].scatter(row_space_C[:, 0], row_space_C[:, 1], row_space_C[:, 2], color='b', edgecolors='black')
ax2[2].set_title('C row-col space')

plt.show()
```
