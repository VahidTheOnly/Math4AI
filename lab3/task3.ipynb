```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib ipympl
```

```python
A = np.matrix([
    [1, 2, 1],
    [2, -1, -1],
    [-1, 1, -2]])

u = np.random.rand(3, 200)

row_space_A = A.T @ u
col_space_A = A @ u

B = np.matrix([
    [1, 2, -3],
    [3, 1, 1],
    [2, 1, 0]])

row_space_B = B.T @ u
col_space_B = B @ u

C = np.matrix([
    [1, 2, -3],
    [3, 6, -9],
    [-2, -4, 6]])

row_space_C = C.T @ u
col_space_C = C @ u
```

```python
fig = plt.figure(figsize=(12,8))
fig.tight_layout()

ax1 = fig.add_subplot(2, 3, 1, projection='3d')
ax4 = fig.add_subplot(2, 3, 4, projection='3d')
ax1.scatter(col_space_A[0, :], col_space_A[1, :], col_space_A[2, :], color='r', edgecolors='black')
ax1.set_title('A col space')
ax4.scatter(row_space_A[0, :], row_space_A[1, :], row_space_A[2, :], color='r', edgecolors='black')
ax4.set_title('A row space')

ax2 = fig.add_subplot(2, 3, 2, projection='3d')
ax5 = fig.add_subplot(2, 3, 5, projection='3d')
ax2.scatter(col_space_B[0, :], col_space_B[1, :], col_space_B[2, :], color='b', edgecolors='black')
ax2.set_title('B col space')
ax5.scatter(row_space_B[0, :], row_space_B[1, :], row_space_B[2, :], color='b', edgecolors='black')
ax5.set_title('B row space')

ax3 = fig.add_subplot(2, 3, 3, projection='3d')
ax6 = fig.add_subplot(2, 3, 6, projection='3d')
ax3.scatter(col_space_C[0, :], col_space_C[1, :], col_space_C[2, :], color='g', edgecolors='black')
ax3.set_title('C col space')
ax6.scatter(row_space_C[0, :], row_space_C[1, :], row_space_C[2, :], color='g', edgecolors='black')
ax6.set_title('C row space')
```

```python
fig = plt.figure(figsize=(12,5))

ax1 = fig.add_subplot(1, 3, 1, projection='3d')
ax1.scatter(col_space_A[0, :], col_space_A[1, :], col_space_A[2, :], color='r', edgecolors='black')
ax1.scatter(row_space_A[0, :], row_space_A[1, :], row_space_A[2, :], color='b', edgecolors='black')
ax1.set_title('A row-col space')

ax2 = fig.add_subplot(1, 3, 2, projection='3d')
ax2.scatter(col_space_B[0, :], col_space_B[1, :], col_space_B[2, :], color='r', edgecolors='black')
ax2.scatter(row_space_B[0, :], row_space_B[1, :], row_space_B[2, :], color='b', edgecolors='black')
ax2.set_title('B row-col space')

ax3 = fig.add_subplot(1, 3, 3, projection='3d')
ax3.scatter(col_space_C[0, :], col_space_C[1, :], col_space_C[2, :], color='r', edgecolors='black')
ax3.scatter(row_space_C[0, :], row_space_C[1, :], row_space_C[2, :], color='b', edgecolors='black')
ax3.set_title('C row-col space')
```
