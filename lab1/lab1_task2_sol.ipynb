```python
import numpy as np
import scipy.io.wavfile
import matplotlib.pyplot as plt

sampling_rate, data = scipy.io.wavfile.read('./requirements/voice1.wav')

print('samping rate:', sampling_rate) # frequency (sample per second)
print('data type:', data.dtype)
print('data shape:', data.shape)

N, no_channels = data.shape # signal length and no. of channels

print('signal length:', N)

channel0 = data[:,0]
channel1 = data[:,1]

scale = 2**np.linspace(-2,4, N)

plt.plot(np.arange(N), scale)
plt.show()

print('shape_old:', scale.shape)
scale.shape = (N,1)
print('shape_new:', scale.shape)

data_new = np.int16(scale * data)
```

```python
plt.figure(figsize=(12, 5))
plt.plot(channel0)
plt.plot(channel1)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(data_new)
plt.show()
```
